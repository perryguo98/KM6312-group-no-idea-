```python
import pandas as pd
import numpy as np
np.random.seed(123)
import matplotlib
matplotlib.rcParams['font.size'] = 8
matplotlib.rcParams['figure.figsize'] = (5, 5)
import seaborn as sns
from IPython.core.pylabtools import figsize
# Scipy helper functions
import scipy
from scipy.stats import percentileofscore
from scipy import stats
```

```python
import pandas as pd
file_name = "result_mat.csv"
df_result = pd.read_csv(file_name, sep=',', index_col=None)
```

```python
df_result['ave_Grade']=(df_result['G1']+df_result['G2'])/2
df_result.head(2).append(df_result.tail(2))
```

```python
file_name = "student-mat.csv"
df_data = pd.read_csv(file_name, sep=',', index_col=None)
```

```python
df_data["Finalc"]=df_data["Dalc"]+df_data["Walc"]
```

```python
df_data['ave_Grade']=(df_data['G1']+df_data['G2'])/2
```

```python
df_data.head(2).append(df_data.tail(2))
```

# Age Distribution 

```python
group = df_result.groupby("age").agg("mean")
```

```python
group
```

```python
#纯年龄分布
plt.bar(df_data['age'].value_counts().index, 
        df_data['age'].value_counts().values,
         color = 'darkcyan', edgecolor = 'w', width = 1,alpha=0.6)
plt.xlabel('age'); 
plt.ylabel('Count'); 
plt.title('Distribution of Ages');

plt.grid(axis='x', which='major',b=False)

x1=df_data['age'].value_counts().index
y1=df_data['age'].value_counts().values

for x,y in zip(x1,y1):
    plt.text(x,y,y,ha='center', va='bottom', fontsize=12,color='xkcd:dark grey')

plt.xticks(list(range(15, 23)));
```

```python
fig,ax = plt.subplots(figsize=(10,5), dpi=80)
ax.set_title("Distribution of Ages by Sex")

sns.countplot(x=df_data['age'],
              hue=df_data['sex'],
              palette="Pastel1",
              data=df_data)

for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}',
                (p.get_x(), p.get_height()),
                color='xkcd:dark grey', size=12)
    

plt.show()
```

```python
#每个年龄的平均喝酒量
plt.bar(group['Finalc'].index, 
        group['Finalc'].values,
         color = 'xkcd:light blue', edgecolor = 'w', width = 1,alpha=0.8)

plt.xlabel('Age'); 
plt.ylabel('Average Alcohol Consumption'); 
plt.title('Distribution of Alcohol Consumption in different age group');

x1=group['Finalc'].index
y1=group['Finalc'].values
for x,y in zip(x1,y1):
    plt.text(x,y,'%.2f' % y,ha='center', va='bottom', fontsize=12,color='xkcd:dark grey')
```

```python
plt.figure(dpi=80)
plt.title('Average Alcohol Consumption of Ages by Sex');

sns.pointplot(x=df_data['age'], y=df_data['Finalc'], 
              hue=df_data['sex'], palette="Pastel1",
              data=df_data,ci=30)
```

```python
#不同年龄的喝酒量/成绩/学习时间
x = range(15,23)
plt.figure(figsize=(10,5),dpi=80)
plt.xlabel("age")
plt.ylabel("Count")
plt.plot(x,group['studytime'].values,label="studytime")
plt.plot(x,group['Finalc'].values,label="Alcohol")
plt.plot(x,group['freetime'].values,label="freetime")
plt.plot(x,group['ave_Grade'].values,label="Grade")
plt.legend()
```

```python
import matplotlib.pyplot as plt
import matplotlib
import seaborn
#matplotlib.rcParams['font.sans-serif'] = ['SimHei']
#matplotlib.rcParams['axes.unicode_minus'] = False

label_list = ["Male", "Female"]   # 各部分标签
size = [187, 208]    # 各部分大小
color = ["skyblue", "lightsalmon"]    # 各部分颜色
explode = [0, 0]   # 各部分突出值
patches, l_text, p_text = plt.pie(size, explode=explode, colors=color, labels=label_list,
                                  labeldistance=1.1, autopct="%1.1f%%", shadow=False, startangle=90, pctdistance=0.6)
plt.axis("equal")    # 设置横轴和纵轴大小相等，这样饼才是圆的
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))

x =np.arange(2)
total_width, n = 0.8, 3

width = total_width / n
x = x - (total_width - width) / 2

plt.bar(x, group_sex['studytime'].values, 
        width=width, label='studytime',
        fc = 'powderblue')

plt.bar(x+width, group_sex['Finalc'].values, width=width, label='Alcohol',
        fc = 'khaki')
    
plt.bar(x + 2*width, group_sex['ave_Grade'].values, width=width, label='Grade',
        fc = 'pink')

x1=x
x2=x+width
x3=x + 2*width
list1=group_sex['studytime'].values
list2=group_sex['Finalc'].values
list3=group_sex['ave_Grade'].values

for a,b in zip(x2,list2):
    plt.text(a,b,'%.2f' % b,ha='center', va='bottom', fontsize=11,color='xkcd:dark grey')

for a,b in zip(x1,list1):
    plt.text(a,b,'%.2f' % b,ha='center', va='bottom', fontsize=11,color='xkcd:dark grey')

for a,b in zip(x3,list3):
    plt.text(a,b,'%.2f' % b,ha='center', va='bottom', fontsize=11,color='xkcd:dark grey')


plt.legend(loc = "best",fontsize=11)
plt.xticks([0,1],['Female','Male'])
plt.show()
```
